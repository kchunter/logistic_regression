```python
def PolynomialFeatures(X1, X2):
    
    if isinstance(X1, (int, float)):
        X1 = np.array([X1])
        X2 = np.array([X2])
    
    degree = 6
    X_polynom = pd.DataFrame()
    
    for i in range(0, degree + 1):
        for j in range(i + 1):
            X_polynom[f"x1^{i - j} * x2^{j}"] = X1**(i-j) * X2**j
    return X_polynom
```

```python
X_polynom = PolynomialFeatures(df['Microchip Test 1'], df['Microchip Test 2'])
```

```python
alpha = 0.001
_lambda = 1000
num_iters = 90000
```

```python
y = df.Accepted
```

```python
y.head()
```

```python
theta = np.zeros((X_polynom.shape[1], 1))
```

```python
def Sigmoid(z):
    return 1/(1 + np.exp(-z))
```

```python
def CostFunction(theta, X, y):
    m = len(y)
    J = 1 / m * (-y.T @ np.log(Sigmoid(X @ theta)) - (1 - y).T @ np.log(1 - Sigmoid(X @ theta))) + _lambda / (2 * m) * (theta[1:, 0].T@theta[1:, 0])
    
    return J[0]
```

```python
def GradientDescent(X, y, theta, alpha, _lambda, num_iters):
    m = len(y)
    J_history = np.zeros((num_iters, 1))
    X0 = X[:, 0][:, None]
    y = y[:, None]
    
    for i in range(num_iters):
    
        theta[0, 0] = theta[0, 0] - alpha * (1 / m) * (X0.T @ (Sigmoid(X @ theta) - y))[0, 0]
        
        theta[1:, 0] = theta[1:, 0] - alpha * (1 / m) * (np.squeeze((X[:, 1:].T @ (Sigmoid(X @ theta) - y))) + _lambda / m * theta[1:, 0])
        
        J_history[i] = CostFunction(theta, X, y)
        
    return theta, J_history
```

```python
theta, J_history = GradientDescent(X_polynom.values, y.values, theta, alpha, _lambda, num_iters)
```

```python
def PlotCost():
    plt.plot(range(num_iters), J_history)
    plt.xlabel('Number of iterations')
    plt.ylabel('Cost')
    plt.title(f'Lambda: {_lambda}')
    plt.show()
```

```python
def CalculateDecisionBoundary():
    u = np.linspace(-1, 1.5, 50)
    v = np.linspace(-1, 1.5, 50)
    z = np.zeros((len(u), len(v)))
    
    for i in range(0, len(u)):
        for j in range(0, len(v)):
            z[i, j] = PolynomialFeatures(u[i], v[j]).values@theta

    z = z.T
    return u, v, z
```

```python
def PlotDecisionBoundary():
    u, v, z = CalculateDecisionBoundary()
    plt.plot(pos['Microchip Test 1'], pos['Microchip Test 2'], 'go', label='Accepted')
    plt.plot(neg['Microchip Test 1'], neg['Microchip Test 2'], 'rx', label='Rejected')
    plt.contour(u, v, z, 0)
    plt.title(f'Lambda {_lambda}')
    plt.xlabel('Microchip Test 1 score')
    plt.ylabel('Microchip Test 2 score')
    plt.legend(loc="upper right")
    plt.show()
```
